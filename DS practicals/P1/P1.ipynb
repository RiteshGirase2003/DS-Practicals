5 - Data Formatting and Data Normalization

i. Data Formatting

```python

# Sample data
data = {
    'ID': [1, 2, 3, 4],
    'DateOfBirth': ['1990-05-15', '1985-10-20', '1978-12-08', '1995-02-28'],
    'Salary': ['$60,000', '$45,000', '$70,000', '$55,000'],
    'IsMarried': [1, 0, 1, 0]
}
df = pd.DataFrame(data)

# Data Formatting
# Convert DateOfBirth to datetime format
print(df.dtypes)
df['DateOfBirth'] = pd.to_datetime(df['DateOfBirth'], format='%Y-%m-%d', errors='coerce')

# Convert Salary to numeric format
df['Salary'] = df['Salary'].replace('[\$,]', '', regex=True).astype(float)

# Display formatted data
print(df)
print(df.dtypes)
```

ii. Data Normalization

```python
from sklearn.preprocessing import MinMaxScaler

# Generate sample data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Height': [160, 175, 180, 165],
    'Gender': ['Female', 'Male', 'Male', 'Male'],
    'IsStudent': [True, False, True, False]
}
df = pd.DataFrame(data)

# Normalization
scaler = MinMaxScaler()
numeric_cols = ['Age', 'Height']
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Display normalized data
print(df)
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, Normalizer

# Generate a small dataset
np.random.seed(0)
data = np.random.randint(0, 100, size=(5, 3))
print("Original Data:")
print(data)

# 1. Min-Max Scaling
# this is Xmin and Xmax is calculated for each column
#  Xnorm = ( X - Xmin ) / ( Xmax - Xmin )

min_max_scaler = MinMaxScaler()
data_min_max = min_max_scaler.fit_transform(data)
print("\nMin-Max Scaled Data:")
print(data_min_max)

Xmin = data.min(axis=0)
Xmax = data.max(axis=0)
print(Xmin,Xmax)
data_min_max1 = (data - Xmin) / (Xmax - Xmin)

print("\nMin-Max Scaled Data1 : ")
data_min_max1 = np.reshape(data_min_max1,(5,3))
print(data_min_max1)

# 2. Z-Score Standardization
#  Xnorm = (X - mean)/std deviation
standard_scaler = StandardScaler()
data_standard = standard_scaler.fit_transform(data)
print("\nZ-Score Standardized Data:")
print(data_standard)

mean = data.mean(axis=0)
std_dev = data.std(axis=0)
data_standard_manual = (data - mean) / std_dev
print("\nManual Z-Score Standardized Data:")
print(data_standard_manual)



# 3. Robust Scaling
robust_scaler = RobustScaler()
data_robust = robust_scaler.fit_transform(data)
print("\nRobust Scaled Data:")
print(data_robust.shape)
print(data_robust)


Q1 = np.percentile(data, 25, axis=0)
Q3 = np.percentile(data, 75, axis=0)
IQR = Q3 - Q1
data_robust_manual = (data - Q1) / IQR
print("\nManual Robust Scaled Data:")
print(data_robust_manual)



# 4. Unit Vector Scaling
normalizer = Normalizer()
data_normalized = normalizer.fit_transform(data)
print("\nUnit Vector Scaled Data:")
print(data_normalized)


norms = np.linalg.norm(data, axis=1, keepdims=True)
data_normalized_manual = data / norms
print("\nManual Unit Vector Scaled Data:")
print(data_normalized_manual)
```
