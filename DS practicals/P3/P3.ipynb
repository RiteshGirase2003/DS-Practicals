```python
import numpy as np
import pandas as pd

data = pd.read_csv('income.csv')

print(data.head())
```

```python
# 1
print(np.var(data['Age']))

print(np.mean(data['Age']))

print(np.median(data['Age']))

print(np.min(data['Age']))

print(np.max(data['Age']))

print(np.std(data['Age']))


data.describe()
```

```python

#2
print(np.var(data['Income']))

print(np.mean(data['Income']))

print(np.median(data['Income']))

print(np.min(data['Income']))

print(np.max(data['Income']))

print(np.std(data['Income']))

```

```python
categorical_variable = 'Age'  
quantitative_variable  = 'Income'         
summary_stats = data.groupby(categorical_variable)[quantitative_variable].describe()
print(summary_stats.head(5).sum())


summary_stats.head(5)
```

2 - IRIS statistical details like percentile, mean, standard deviation etc

```python
import pandas as pd

# Load the iris dataset
iris_data = pd.read_csv('IRIS.csv')

# Filter data for each species
setosa_data = iris_data[iris_data['species'] == 'Iris-setosa']
versicolor_data = iris_data[iris_data['species'] == 'Iris-versicolor']
virginica_data = iris_data[iris_data['species'] == 'Iris-virginica']

# Basic statistical details for each species
setosa_stats = setosa_data.describe()
versicolor_stats = versicolor_data.describe()
virginica_stats = virginica_data.describe()

# Display statistical details
print("Statistical Details for Iris-setosa:")
print(setosa_stats)
print()

print("Statistical Details for Iris-versicolor:")
print(versicolor_stats)
print()

print("Statistical Details for Iris-virginica:")
print(virginica_stats)
```
