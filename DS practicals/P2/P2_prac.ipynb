```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('tecdiv.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
# To handle missing values 
from sklearn.impute import SimpleImputer
df = data.copy()
df.columns = df.columns.str.strip()
col_to_drop = ['Timestamp','Email Address','Name','Email','Roll_no','PRN No.','Mobile No.']
df = df.drop(columns=col_to_drop)
```

```python
mean_imputer = SimpleImputer(strategy='mean')
imputed_df = pd.DataFrame(mean_imputer.fit_transform(df),columns=df.columns)
imputed_df
```

```python
median_imputer = SimpleImputer(strategy='median')
imputed_df = pd.DataFrame(median_imputer.fit_transform(df),columns=df.columns)
imputed_df
```

```python
most_imputer = SimpleImputer(strategy='most_frequent')
imputed_df = pd.DataFrame(most_imputer.fit_transform(df),columns=df.columns)
imputed_df
```

```python
const_imputer = SimpleImputer(strategy='constant', fill_value=11)
imputed_df= pd.DataFrame(const_imputer.fit_transform(df),columns=df.columns)
imputed_df
```

```python
data = {
    'Student_ID': np.arange(1, 11),
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily', 'Frank', 'Grace', 'Hannah', 'Ian', 'Julia'],
    'Math_Score': [85, 95, 82, 91, 2, 1, 67, 57, 82, 98],
    'Science_Score': np.random.randint(50, 100, 10),
    'English_Score': np.random.randint(50, 100, 10),
    'History_Score': np.random.randint(50, 100, 10)
}
df_3 = pd.DataFrame(data)

Q1 = df_3.quantile(0.25)
Q3 = df_3.quantile(0.75)

IQR = Q3 - Q1
outliers = ( (df_3 < (Q1 - 1.5*IQR) ) | (df_3 > (Q3 + 1.5*IQR))).any(axis=1)
outliers
```

```python
df_3[outliers]
```

```python
df_3 = df_3[~outliers]
df_3
```
